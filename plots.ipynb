```python
df['timestamp'] = pd.to_datetime(df['date_time'])
tmp = df.groupby('srch_id').first().sort_values('date_time')
# Plotting
# Extract year and month and create a new column 'year_month'
tmp['date_time'] = pd.to_datetime(tmp['date_time'])

tmp['year_month'] = tmp['date_time'].dt.to_period('W')

weekly_search_counts = tmp.groupby('year_month').size()

# Convert the index to string for plotting
weekly_search_counts.index = weekly_search_counts.index.astype(str)

# Plot the data as a line plot
plt.figure(figsize=(10, 6))
plt.plot(weekly_search_counts.index, weekly_search_counts.values, marker='o')
plt.xlabel('Week')
plt.ylabel('Number of Searches')
plt.title('Number of Searches Per Week Train Set')
plt.xticks([])  # Hide x-axis labels

plt.tight_layout()

plt.savefig('./plots/srch_freq_week_train')

plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame df with columns click_bool and booking_bool

# Filter the DataFrame for clicked properties
clicked_properties = df[df['click_bool'] == 1]

# Count the occurrences of booking_bool for clicked properties
booking_counts = clicked_properties['booking_bool'].value_counts()

# Plotting
plt.figure(figsize=(8, 6))
bars = plt.bar(booking_counts.index.astype(str), booking_counts.values, color=['skyblue', 'salmon'])

# Add labels and title
plt.xlabel('Booking Status')
plt.ylabel('Count')
plt.title('Distribution of Clicked Properties and Booking Status')
plt.xticks(range(len(booking_counts)), ['Not Booked', 'Booked'])
plt.ylim(0, max(booking_counts.values) * 1.1)  # Adjust ylim to provide space for labels

# Add counts on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, yval, ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.savefig("./plots/clicked_booked_dist")

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming you have a DataFrame `df` with 15 columns
# Example: df = pd.DataFrame({
#     'col1': [...],
#     'col2': [...],
#     ...
#     'col15': [...]
# })

# Number of columns
num_columns = 4
cols = df.columns[:num_columns]

# Set the number of rows and columns for the subplot grid
nrows = 5
ncols = 3

fig, axes = plt.subplots(nrows, ncols, figsize=(15, 20))
fig.tight_layout(pad=5.0)

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Iterate through each column and plot the bar plot
for i, col in enumerate(cols):
    sns.countplot(data=df, x=col, ax=axes[i])
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')

# Remove any extra subplots
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

# Display the plot
plt.show()
```
